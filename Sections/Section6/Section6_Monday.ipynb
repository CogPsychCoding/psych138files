## Review of Functions with Input Parameters
- input parameters (or input arguments)
- type hints
- defining default arguments
    - define default value
    - optional parameters (or arguments)
- using only a subset of default arguments



````python
def identifying_info(firstname: str, lastname: str="smith"):
    info = f"This person is {firstname} {lastname}."  # we can use the variable names defined in input arguments to use the input values throughout our function
    lastname = lastname + 'change'

    return info


rick_info = identifying_info("rick")  # call with only one input argument
print(rick_info)  # lastname is optional, if no value is input for lastname during call, then the function will use the default value

rick_info = identifying_info(55)  # runs fine without matching type-hint
print(rick_info)

rick_info = identifying_info("rick", "james")  # call with two input arguments
print(rick_info)  # a lastname value is input during call, so the function will not use the default value

rick_info = identifying_info(firstname="rick", lastname="billings")  # we can use "keyword arguments", where we assign a value to the relevant variable name
print(rick_info)

rick_info = identifying_info(lastname="billings", firstname="rick")  # if we do so, then location does not matter.
print(rick_info)

lastname = "something_different"
rick_info = identifying_info("billings", "rick")  # in otherwords, if we dont use keywords, location DOES matter
print(rick_info)
print(lastname)
````

````python
def identifying_info(firstname: str, lastname: str="smith", age: int=35, location: str="nyc"):
    info = f"This person is {firstname} {lastname}, aged {age}, from {location}"

    return info


# if we are CAREFUL with LOCATION, then we don't need to use keyword arguments
lewis_info = identifying_info("lewis", "ham", '22', "uk")
print(lewis_info)

# but keyword arguments can your function call more informative and easier to read
lewis_info = identifying_info("lewis", lastname="ham", age='22', location="uk")
print(lewis_info)

# keyword arguments let us change the values of chosen input arguments, and we can skip LOCATION, if the skipped input arguments have default values
lewis_info = identifying_info("lewis", location="uk")
print(lewis_info)
````
