# Why Python?

C Code

```python
#include <stdio.h>
#include <stdlib.h>

int main() {
    // Create an array of strings containing the words "one" through "nine"
    char *words[] = {"one", "two", "three", "four", "five", "six", "seven", "eight", "nine"};

    // Calculate the number of words in the array
    int numWords = sizeof(words) / sizeof(words[0]);

    // Loop through the array and print each word
    for (int i = 0; i < numWords; i++) {
        printf("%s\n", words[i]);
    }

    return 0;
}
```

C++ Code

```python
#include <iostream>
#include <vector>
#include <string>

int main() {
    // Create a vector of strings containing the words "one" through "nine"
    std::vector<std::string> words = {"one", "two", "three", "four", "five", "six", "seven", "eight", "nine"};

    // Loop through the vector and print each word
    for (const std::string& word : words) {
        std::cout << word << std::endl;
    }

    return 0;
}
```

Java Code

```python
public class Main {
    public static void main(String[] args) {
        // Create an array of strings containing the words "one" through "nine"
        String[] words = {"one", "two", "three", "four", "five", "six", "seven", "eight", "nine"};

        // Loop through the array and print each word
        for (String word : words) {
            System.out.println(word);
        }
    }
}
```
